```python
import pandas as pd
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
from itertools import combinations
import numpy as np
df = pd.read_csv("C:/Users/pawithlocal/pawith\program_coding/KUxDEPA2022/Datasets/Telco-Customer-Churn.csv")
column_object = [i for i in df.columns if i not in ['SeniorCitizen', 'tenure', 'MonthlyCharges', 'TotalCharges',
                                                    'Churn', 'total_usage_month','customerID']]
X = df[column_object]
y = df[["Churn"]]
y.info()
X_train, X_test, y_train,  y_test = train_test_split(X, y, 
                                                    test_size=0.2, 
                                                    train_size=0.8)
import warnings

with warnings.catch_warnings():
    # this will suppress all warnings in this block
    warnings.simplefilter("ignore")
    # compute combination of n columns that select to be features
    
    n_col_select = 2
    comb = list(combinations([i for i in range(0,len(column_object),1)], n_col_select))
    comb = np.array(comb)
    col_object_np = np.array(column_object)
    train_score = []
    test_score = []
    col_index = []
    for col in comb:
        col_selected = list(col_object_np[col])  # columns that have selected

        transformer = make_column_transformer(
            (StandardScaler(), []),
            (OneHotEncoder(), col_object_np[col]),
            remainder = "passthrough")

        X_transformed_train = transformer.fit_transform(X_train[col_selected])
        X_transformed_test = transformer.fit_transform(X_test[col_selected])
        n = 14
        knn = KNeighborsClassifier(n_neighbors=n)
        knn.fit(X_transformed_train, y_train)
        train_score.append(knn.score(X_transformed_train, y_train))
        test_score.append(knn.score(X_transformed_test, y_test))

        col_index.append(col_selected)
a = pd.DataFrame({"column":col_index, "train_score":train_score, "test_score":test_score})
a.sort_values(["test_score"], ascending=False)
a
```
